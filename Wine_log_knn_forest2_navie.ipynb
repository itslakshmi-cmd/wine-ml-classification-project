```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('wine.data.csv')
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
y=df['Class']
x=df.drop(columns='Class',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
log_wine=LogisticRegression()
log_wine.fit(x_train,y_train)
```

```python
y_pred_log=log_wine.predict(x_test)
y_pred_log[:5]
```

```python
y_test[:5]
```

```python
from sklearn.metrics import accuracy_score
Log_wine_Acc=accuracy_score(y_pred_log,y_test)
```

```python
Log_wine_Acc
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
neighbors=np.arange(1,11)
train_Acc=[]
test_Acc=[]
for i in neighbors:
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    train_Acc.append(knn.score(x_train,y_train))
    test_Acc.append(knn.score(x_test,y_test))
```

```python
import matplotlib.pyplot as plt
plt.plot(neighbors,train_Acc,test_Acc)
labels=["Testing_accuracy","Training_accuracy"]
```

```python
knn_wine=KNeighborsClassifier(n_neighbors=4)
knn_wine.fit(x_train,y_train)
```

```python
y_pred_knn=knn_wine.predict(x_test)
```

```python
y_pred_knn[:3]
```

```python
y_test[:3]
```

```python
from sklearn.metrics import accuracy_score
knn_wine_acc=accuracy_score(y_pred_knn,y_test)
```

```python
knn_wine_acc
```

Navie Bayes

```python
from sklearn.naive_bayes import GaussianNB
Naive_wine=GaussianNB()
Naive_wine.fit(x_train,y_train)
```

```python
y_pred_Naive=Naive_wine.predict(x_test)
```

```python
y_pred_Naive[:3]
```

```python
y_test[:5]
```

```python
from sklearn.metrics import accuracy_score
Naive_wine_Acc=accuracy_score(y_pred_Naive,y_test)
```

```python
Naive_wine_Acc
```

Decision_Tree

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(max_leaf_nodes=10,criterion='entropy')
model.fit(x_train,y_train)
```

```python
y_pred_Tree=model.predict(x_test)
y_pred_Tree[:3]
```

```python
from sklearn.metrics import accuracy_score
Tree_wine_acc=accuracy_score(y_pred_Tree,y_test)
Tree_wine_acc
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(33,20))
plot_tree(model,feature_names=x.columns,filled=True,fontsize=15)
```

```python
models=['LogisticRegression','KNN','Decision_Tree','Naive']
accuracy_score=[Log_wine_Acc,knn_wine_acc,Tree_wine_acc,Naive_wine_Acc]
a=pd.DataFrame({'model':models,'accuracy':accuracy_score})
```

```python
import seaborn as sns
sns.barplot(a,x='model',y='accuracy')
```

Project Summary: Wine Classification using Machine Learning

This project focuses on classifying wine samples into different classes based on their chemical properties using multiple machine learning algorithms. The dataset was explored, cleaned, and split into training and testing sets before applying various classification models.

The following algorithms were implemented and evaluated:

Logistic Regression

K-Nearest Neighbors (KNN)

Naive Bayes

Decision Tree

Random Forest

Model performance was measured using accuracy score, and KNN performance was further analyzed by tuning the number of neighbors and visualizing training vs testing accuracy.

Key Outcomes

The dataset contained no missing values, making preprocessing straightforward.

Random Forest and KNN showed strong performance compared to other models.

KNN model tuning helped identify the optimal number of neighbors.

Ensemble methods like Random Forest improved classification accuracy over simpler models.
